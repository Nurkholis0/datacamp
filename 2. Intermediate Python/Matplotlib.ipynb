## Build a Histogram (2): bins

In the previous exercise, you didn't specify the number of bins. By default, **Python sets the number of bins to 10 in that case**. $\color{red}{\text{The number of bins}}$ is pretty important. Too few bins will oversimplify reality and won't show you the details. **Too many bins will overcomplicate reality and won't show the bigger picture**.

To **control the number of bins to divide your data in**, you can set $\color{blue}{\text{the bins argument}}$.

That's exactly what you'll do in this exercise. You'll be making two plots here. $\color{blue}{\text{plt.show()}}$ displays a plot; $\color{blue}{\text{plt.clf()}}$ cleans it up again so you can start afresh.

```python
# import numpy packages
import numpy as np

# import pandas
import pandas as pd

# import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# read csv file: gdp_csv
gdp_csv = pd.read_csv("gapminder.csv")

# create numpy arrays for csv files: np_gdp_csv
np_gdp_csv = np.array(gdp_csv)

# create life_exp for  life expectancy for each country in 2007
life_exp = np_gdp_csv[:,5]

# Build histogram with 5 bins
plt.hist(life_exp, bins=5, edgecolor="blue", color="red")

# Show and clean up plot
plt.show()
plt.clf()

# Build histogram with 20 bins
plt.hist(life_exp, bins=20, edgecolor="red", color="orange")

# Show and clean up again
plt.show()
plt.clf()
```

## Labels

It's time to customize your own plot. This is the fun part, you will see your plot come to life!

You're going to work on the scatter plot with world development data: GDP per capita on the x-axis (logarithmic scale), life expectancy on the y-axis.

As a first step, let's add axis labels and a title to the plot. You can do this with the $\color{blue}{\text{xlabel()}}$, $\color{blue}{\text{ylabel()}}$ and $\color{blue}{\text{title()}}$ functions, available in $\color{blue}{\text{matplotlib.pyplot}}$.

```python
# import numpy packages
import numpy as np

# import pandas packages
import pandas as pd

# import matplotlib packages
import matplotlib.pyplot as plt

# read CSV file with pandas packages
read_csv = pd.read_csv("gapminder.csv")

# create np_read_csv to make list numpy arrays
np_read_csv = np.array(read_csv)

# create list gdp_cap for GDP per Capita [in USD]
gdp_cap = list(np_read_csv[:, 6])

# create list life_exp for Life Expectancy [in years]
list_exp = list(np_read_csv[:, 5])

# Basic scatter plot, log scale
plt.scatter(gdp_cap, life_exp)
plt.xscale('log')

# Strings for xlab, ylab, and title
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'

# create plt.xlabel, plt.ylabel, and plt.title for add axis label and title
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)

# create plt.show() for display the plot
plt.show()
```

## Sizes

Right now, the scatter plot is just a cloud of blue dots, indistinguishable from each other. Let's change this. Wouldn't it be nice if the size of the dots corresponds to the population?

To accomplish this, there is a list $\color{blue}{\text{pop}}$ loaded in your workspace. It contains population numbers for each country expressed in millions. You can see that this list is added to the scatter method, as the argument $\color{blue}{\text{s}}$, for size.

```python
# import numpy packages
import numpy as np

# import pandas packages
import pandas as pd

# import matplotlib packages
import matplotlib.pyplot as plt

# create_csv for open csv file
create_csv = pd.read_csv("gapminder.csv")

# create list with numpy arrays: np_create_csv
np_create_csv = np.array(create_csv)

# create list gdp_cap for GDP per Capita [in USD]
gpd_cap = list(np_create_csv[:,6])

# create list life_exp for Life Expectancy [in years]
life_exp = list(np_create_csv[:,5])

# create list pop for population
pop = list(np_create_csv[:,3] * 0.000001)

# Store pop as a numpy array: np_pop
np_pop = np.array(pop)

# Double np_pop
np_pop = np_pop*2

# Scatter plot for x-axis is gdp_cap, y-axis is life_exp and s for size  
plt.scatter(gdp_cap, life_exp, s = np_pop)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])

# After customizing, display the plot
plt.show()
```

## Colors

The next step is making the plot more colorful! To do this, a list $\color{blue}{\text{col}}$ has been created for you. It's a list with a color for each corresponding country, depending on the continent the country is part of.

How did we make the list $\color{blue}{\text{col}}$ you ask? The Gapminder data contains a list continent with the $\color{blue}{\text{continent}}$ each country belongs to. A dictionary is constructed that maps continents onto colors:

```python
dict = {
    'Asia':'red',
    'Europe':'green',
    'Africa':'blue',
    'Americas':'yellow',
    'Oceania':'black'
}
```

Nothing to worry about now; you will learn about dictionaries in the next chapter.

```python
# import numpy packages
import numpy as np

# import pandas packages
import pandas as pd

# import matplotlib packages
import matplotlib.pyplot as plt

# create_csv for open csv file
create_csv = pd.read_csv("gapminder.csv")

# create list with numpy arrays: np_create_csv
np_create_csv = np.array(create_csv)

# create list gdp_cap for GDP per Capita [in USD]
gpd_cap = list(np_create_csv[:,6])

# create list life_exp for Life Expectancy [in years]
life_exp = list(np_create_csv[:,5])

# create list pop for population
pop = list(np_create_csv[:,3] * 0.000001)

#create color_txt for open color.txt
color_txt = list(np.loadtxt("col.txt", dtype=str))

# Specify c and alpha inside plt.scatter()
plt.scatter(gdp_cap, life_exp, s = np.array(pop) * 4, c = color_txt, alpha=0.5)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])

# After customizing, display the plot
plt.show()
```
